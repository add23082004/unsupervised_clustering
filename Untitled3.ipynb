```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns


# Load your dataset
mdset = pd.read_csv("Credit_card_dataset.csv")
```

```python
# Calculate the mean of the variable CREDIT_LIMIT
mean_value = mdset['CREDIT_LIMIT'].mean()

# Replace missing values with the mean
mdset['CREDIT_LIMIT'].fillna(mean_value, inplace=True)

# Drop useless variables
mdset.drop('CUST_ID', axis=1, inplace=True)
```

```python
# Select relevant features
selected_features = mdset[['CASH_ADVANCE', 'PAYMENTS', 'BALANCE_FREQUENCY', 'PURCHASES', 'CREDIT_LIMIT']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(selected_features)

# Apply Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
mdset['hierarchical_cluster'] = hierarchical.fit_predict(scaled_features)

# Visualize Hierarchical Clustering Results
plt.figure(figsize=(7, 5))
sns.scatterplot(x='CASH_ADVANCE', y='PAYMENTS', hue='hierarchical_cluster', data=mdset, palette='viridis')
plt.title('Hierarchical Clustering Results')
plt.show()
```

```python
# Apply K-means clustering (with an arbitrary initial choice of k)
kmeans = KMeans(n_clusters=4, random_state=42)
mdset['kmeans_cluster'] = kmeans.fit_predict(scaled_features)

# Visualize K-means Clustering Results
plt.figure(figsize=(7, 5))
sns.scatterplot(x='CASH_ADVANCE', y='PAYMENTS', hue='kmeans_cluster', data=mdset, palette='viridis')
plt.title('K-means Clustering Results (Arbitrary k)')
plt.show()
```

```python
# Calculate elbow point for K-means clustering
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum_of_squarred_errors')
plt.title('Elbow Method for K-means Clustering')
plt.show()
```

```python
# Choose the optimized k value based on the elbow method
optimal_k = 3 

# Reapply K-means clustering with the optimized k value
kmeans_optimized = KMeans(n_clusters=optimal_k, random_state=42)
mdset['kmeans_optimized_cluster'] = kmeans_optimized.fit_predict(scaled_features)

# Visualize the optimized K-means clustering results
plt.figure(figsize=(7, 5))
sns.scatterplot(x='PURCHASES', y='CREDIT_LIMIT', hue='kmeans_optimized_cluster', data=mdset, palette='viridis')
plt.title('Optimized K-means Clustering Results')
plt.show()
```
